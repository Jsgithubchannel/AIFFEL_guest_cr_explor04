```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.price_sum = 0  # 총 가격 초기화
        self.order_menu = [] # 주문한 음료 리스트
        self.order_price = [] # 주문한 가격 리스트

# 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

# 주문 매서드
    def menu_select(self):
        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력
            if 1 <= n and n <= len(self.menu):
                self.price_sum += self.price[n - 1]  # 선택 음료의 가격 추가

                # 음료 온도 물어보기
                t = 0  # 기본값을 넣어주고
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.")

                # 주문 리스트에 추가 (온도 포함)
                self.order_menu.append(f"{self.temp} {self.menu[n - 1]}")
                self.order_price.append(self.price[n - 1])  # 가격 리스트에 추가

                print(f'{self.temp} {self.menu[n - 1]} : {self.price_sum}원')
            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if 1 <= n <= len(self.menu):  # 추가 주문
                self.price_sum += self.price[n - 1]  # 추가 음료 가격 합산
                # 추가 음료 온도 입력받기
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.")
                self.order_menu.append(f"{self.temp} {self.menu[n - 1]}")  # 주문 리스트에 추가 (온도 포함)
                self.order_price.append(self.price[n - 1])  # 가격 리스트에 추가
                print(f'추가 주문 음료 {self.temp} {self.menu[n - 1]} : {self.price[n - 1]}원\n합계 : {self.price_sum}원')

            elif n == 0:  # 지불을 입력하면

                print("주문이 완료되었습니다.")
                print("주문 내역:", self.order_menu)
                print("각 음료 가격:", self.order_price)
                print("총 합계:", self.price_sum, "원")
            else:  # 없는 번호를 입력할 때
                print("없는 메뉴입니다. 다시 주문해 주세요.")
    def pay(self):
        while True:
          print("결제 방법을 선택하세요:")
          print("1. 현금(cash)")
          print("2. 카드(card)")
          payment_method = input("결제 방법을 입력하세요 (1: cash, 2: card): ").strip().lower()

          if payment_method in ['cash', '1']:
            print("직원을 호출하겠습니다.")
            break
          elif payment_method in ['card', '2']:
            print("IC칩 방향에 맞게 카드를 꽂아주세요.")
            break
          else:
            print("올바른 결제 방법을 선택하세요.")

    def table(self):
      print('⟝' + '-' * 30 + '⟞')
      for i in range(5):
        print('|' + ' ' * 31 + '|')
      for i in range(len(self.order_menu)):
        print(self.order_menu[i], ' : ', self.order_price[i])
      print('합계 금액 :', self.price_sum)
      for i in range(5):
        print('|' + ' ' * 31+ '|')
      print('⟝' + '-' * 30 + '⟞')

a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()
a.table()
```
