```python
import random

# 캐릭터 클래스
class Character:
    def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        reduced_damage = max(0, damage - self.defense)
        self.health = max(0, self.health - reduced_damage)

    def attack_target(self, target):
        damage = random.randint(1, self.attack_power)
        target.take_damage(damage)
        return damage

# 플레이어 클래스
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack_power=25, defense=5)
        self.experience = 0

    def gain_experience(self, amount):
        self.experience += amount
        if self.experience >= 50:
            self.level_up()

    def level_up(self):
        self.level += 1
        self.attack_power += 10
        self.defense += 5
        self.experience = 0
        print(f"{self.name} 레벨업! 레벨: {self.level}, 공격력: {self.attack_power}, 방어력: {self.defense}")

# 몬스터 클래스
class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack_power, defense)

# 배틀 함수 만들기
def battle(player, monster):
    print(f"{monster.name}과의 전투를 시작합니다.")
    while player.is_alive() and monster.is_alive():
        damage_to_monster = player.attack_target(monster)
        print(f"{player.name}이 {monster.name}에게 {damage_to_monster} 만큼 공격했다...!")
        print(f"{monster.name}의 체력: {monster.health}")

        if not monster.is_alive():
            print(f"{monster.name}을 이겼다!")
            player.gain_experience(monster.level * 20)
            return "win"

        damage_to_player = monster.attack_target(player)
        print(f"{monster.name}가 {player.name}에게 {damage_to_player} 만큼 공격했다...!")
        print(f"{player.name}의 체력: {player.health}")

        if not player.is_alive():
            print(f"{player.name}가 졌다...")
            return "lose"

# 플레이 게임 함수 만들기
def play_game():
    monster_dict = {"슬라임": 1, "고블린": 2, "오크": 3}

    player_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        result = battle(player, monster)

        if result == "lose":
            print("게임오버!")
            break

    if player.is_alive():
        print("승리!")

play_game()
```
